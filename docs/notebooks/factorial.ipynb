# Factorial

```python
!source <(curl -s https://raw.githubusercontent.com/conjure-cp/conjure-notebook/v0.0.8/scripts/install-colab.sh)
%load_ext conjure
```

You can use ! after an integer, to get the factorial (product of 1..n) of the integer.

However it can't be used directly in decision variables as shown below.

```python
%%conjure --number-fo-solutions=all
find x: int (1..5) such that x! < 200
```

but can be used as demonstrated below instead.

```python
%%conjure --number-of-solutions=all
find z : int(-1..13)
such that (exists x : int(1..5) . (x! < 200) /\ (z=x))
```
